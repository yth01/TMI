```python
from preprocess import *
```

```python
data = pd.read_csv("../jobkorea_all.csv")
data.columns
```

```python
# 질문, 답변, 조언
data["답변"] = preprocess_main(data["답변"])
data.loc[0, "답변"]
```

```python
# 스펙
data["스펙"] = preprocess_spec(data["스펙"])
data.loc[0, "스펙"]
```

```python
# 스펙 - (1) 학력
data["스펙"].apply(lambda x: x[0]).value_counts()
```

```python
# 스펙 - (2) 전공
data["스펙"][data["스펙"].apply(lambda x: len(x)) > 1].apply(lambda x: x[1]).value_counts()
```

```python
data["스펙"][data["스펙"].apply(lambda x: len(x)) > 2].apply(lambda x: x[2][0:3]).value_counts()
```

```python
data["스펙"][data["스펙"].apply(lambda x: len(x)) > 3].apply(lambda x: x[3][0:3]).value_counts()
```

```python
# 회사명
data["회사명"] = preprocess_company(data["회사명"])
data["회사명"].value_counts()
```

```python
# 직무분야
data["직무분야"] = preprocess_field(data["직무분야"])
data["직무분야"].value_counts()
```
