```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import re
from konlpy.tag import Hannanum
plt.rc('font', family='Malgun Gothic')
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
%matplotlib inline
```

```
df = pd.read_csv("/content/drive/My Drive/TMI/jobkorea_all.csv")
```

```
df_drop = df.columns[1:]
df = df.drop(df_drop, axis=1)
df
```

```
hannanum = Hannanum()
df_pos = df["질문"].apply(lambda x: hannanum.pos(x))
```

```
df_pos.head()
```

```
df_pos = df_pos.apply(lambda x: list(filter(lambda x: x[1] == "N", x)))
```

```
df_pos = df_pos.apply(lambda x: list(map(lambda x: x[0], x)))
df_pos
```

```
df_pos
```

```
data = pd.DataFrame({'질문':df_pos.values})
```

```
data['질문']
```

```
data["질문"] = data["질문"].astype(str)
```

```
data.loc[0, "질문"]
```

```
data["질문"] = data.loc[data["질문"].notna(),"질문"].apply(lambda x: re.sub("\,", "", x))
data["질문"] = data.loc[data["질문"].notna(),"질문"].apply(lambda x: re.sub("\'", "", x))
data["질문"] = data.loc[data["질문"].notna(),"질문"].apply(lambda x: re.sub("[A-Za-z]+", "", x))
data["질문"] = data.loc[data["질문"].notna(),"질문"].apply(lambda x: re.sub("[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]", "", x))
data["질문"]
```

```
list_pos = data["질문"].tolist()
list_pos
```

```
str_pos = ' '.join(list_pos)
str_pos
```

```
list_str = str_pos.split()
series = pd.Series(list_str)
result = series.value_counts()
result
```

```
result = pd.DataFrame({'키워드': result.index, '빈도수':result.values})
result
```
