### 활성화 함수(Activation Function)

딥러닝 네트워크에서는 노드에 들어오는 값들에 대해 곧바로 다음 레이어로 전달하지 않고 주로 비선형 함수를 통과시킨 후 전달한다. 

이때 사용하는 함수를 **활성화 함수(Activation Function)** 이라 부른다.

선형적인 구조만으로 XOR같은 비선형 분류 문제를 해결할 수가 없다. 그렇기 때문에 **활성화 함수**를 사용하여 비선형적인 구조로 변환할 필요가 있다. 또한 선형함수로 신경망을 구성하게 되면 층이 깊어지는 의미가 없이 은닉층이 없는 네트워크로만 표현되게 된다. 따라서 신경망에서 층을 쌓는 혜택을 얻고 싶다면 **활성화 함수**로 반드시 비선형 함수를 사용해야 한다.

활성화 함수의 종류에는 sigmoid, tanh, relu 등 여러 종류의 활성화 함수가 있다. 가장 간단한 활성화 함수로는 sigmoid 함수가 있는데 sigmoid 함수는 Gradient Vanishing와 같은 문제점이 있어 이런 문제를 해결하기 위해 relu 함수를 사용하기도 한다. 하지만 모든 경우 relu 함수가 sigmoid 함수보다 성능이 좋은 것은 아니다. 따라서 모델에 맞는 활성화 함수를 사용할 필요가 있다.

### MNIST 분류

데이터베이스(MNIST : ModifiedNational Institute of Standards and Technologydatabase)는 손으로 쓴 숫자들로 이루어진 대형데이터베이스이며, 다양한화상 처리시스템을트레이닝하기 위해 일반적으로 사용된다. 위에 정리된 것들을 활용하여 MNIST의 숫자 데이터를 보고 모델의 정확도를 올리는 방법을 모색해보자.

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Flatten
from keras.optimizers import SGD, adam
import numpy as np
print(keras.__version__)
```

```python
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

**전처리**

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

**모델 구성(784-128-10로 구성된 신경망)**

```python
model = keras.Sequential([                         
    #  2차원 배열(28 x 28 픽셀)의 이미지 포맷을 28 * 28 = 784 픽셀의 1차원 배열로 변환합니다. 이 층은 이미지에 있는 픽셀의 행을 펼쳐서 일렬로 늘립니다.
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])
```

**모델 컴파일**

```python
model.compile(optimizer=keras.optimizers.SGD(lr=0.01),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**모델 훈련**

```python
hist = model.fit(train_images, train_labels, batch_size=32, epochs=5, verbose=2)
```

**정확도 평가**

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('테스트데이터 정확도:', test_acc)
```

### 통계 기반 언어 모델

**언어 모델(language model)** : 단어 시퀸스에 확률을 부여하는 모델

단어의 등장 순서를 무시하는 **백오브워즈**와 달리 언어 모델은 시퀸스 정보를 명시적으로 학습한다.

~~~
누명을 쓰다 -> 0.41
누명을 당하다 -> 0.02
두시 삼십이분 -> 0.51
이시 서른두분 -> 0.08
난폭 운전 -> 0.39
무모 운전 -> 0.01
~~~


위의 예처럼 자연스러운 한국어 문장에 높은 확률 값을 부여한다,

**n-gram** : n개의 단어를 뜻하는 용어

**2-gram(바이그램)** : [난폭, 운전], [눈, 뜨다]
<br/>**3-gram(트라이그램)** : [누명, 을, 쓰다]
<br/>**4-gram(트라이그램)** : [바람, 잘, 날, 없다]

경우에 따라서 n-gram은 n-gram에 기반한 언어 모델을 의마하기도 한다. 말뭉치 내 단어들을 n개씩 묶어서 그 빈도를 학습했다는 뜻이다.

| 표현 | 빈도 |
|------|------|
| 내   | 1309 |
| 마음 | 172 |
| 속에 | 155 |
|영원히| 104 |
| 기억될   | 29 |
| 최고의   | 3503 |
| 명작이다   | 298 |
| 내 마음   | 93 |
| 마음 속에   | 9 |
| 속에 영원히   | 7 |
| 영원히 기억될   | 7 |
| 기억될 최고의   | 1 |
| 최고의 명작이다   | 23 |
| 영원히 기억될 최고의 명작이다   | 1 |
| 내 마음 속에 영원히 기억될 최고의 명작이다  | 0 |


위 표는 네이버 영화 리뷰 말뭉치에서 각각의 표현이 등장한 횟수를 가르킨다. 띄어쓰기 단위인 어절을 하나의 단어로 보고 빈도를 센 것이다. 마지막 "내 마음 속에 영원히 기억될 최고의 명작이다"라는 말뭉치는 한번도 등장하지 않아서 확률을 0으로 부여하게 된다. 즉, **문법적으로나 의미적으로나 결함이 없는 문장임에도 해당 표현을 말이 되자 않는 문장으로 취급한다는 얘기이다.**

"내 마음 속에 영원히 기억될 최고의"라는 표현 다음에 "명작이다"라는 단어가 나타날 확률은 **조건부확률**의 정의를 활용해 **최대우도측정법**으로 유도하면 밑의 수식과 같다. 하지만 밑의 식은 우변의 분자가 0이기 떄문에 전체 값이 0이 된다.

=$Freq(내, 마음, 속에, 영원히, 기억될, 최고의, 명작이다) \over Freq(내, 마음, 속에, 영원히, 기억될, 최고의)$

n-gram 모델을 쓰면 이런 문제를 일부 해결할 수 있다. 직전 n-1개 단어의 등장 확률로 전체 단어 시퀸스 등장 확률을 **근사**하는 것이다. 이는 한 **상태**의 확률은 그 직전 상태에만 의존한다는 **마코프 가정**에 기반한 것이다. 위의 식을 **바이그램 근사**로 표현하면 밑의 수식과 같다.

=$Freq(최고의, 명작이다) \over Freq(최고의)$  =  $23 \over 3503$

다시 말하자면 "명작이다" 직전의 1개 단어만 보고 전체 단어 시퀸스 등장 확률을 근사한 것이다. 그렇다면 바이그램 모델에서 "내 마음 속에 영원히 기억될 최고의 명작이다"라는 단어 시퀸스가 나타날 확률은 얼마나 될까?

= $1309 \over |V|$ $93 \over 1309$ $9 \over 172$ $7 \over 155$ $7 \over 104$ $1 \over 29$ $23 \over 3503$

하지만 데이터가 한 번도 등장하지 않는 n-gram이 존재할 때 예측 단계에서 문제가 발생할 수 있다. 이를 위해 **백오프(back-off)**, **스무딩(smoothing)** 등의 방식이 제안됐다.

**백오프(back-off)** : n-gram 등장 빈도를 n보다 작은 범위의 단어 시퀸스 빈도로 근사하는 방식
<br/>n을 크게 하면 할수록 등증하지 않은 케이스가 많아질 가능성이 높기 때문이다. 빈도가 1이상인 7-gram에 대해서는 백오프하지 않고 해당 빈도를 그대로 n-gram 모델 학습에 사용한다.

**스무딩(smoothing)** : 위의 표와 같이 같은 등장 빈도 표에 모두 k만큼을 더하는 기법이다.
<br/>이렇게 되면 "내 마음 속에 영훤히 기억될 최고의 명장이다"의 빈도가 k가 된다. 이 떄문에 **Add-k스무딩**이라고 부르기도 한다. 만약 k를 1로 설정한다면 이를 특별히 **라플라스 스무딩**이라고 한다. 스무딩을 시행하면 높은 빈도를 가진 문자열 등장 확률을 일부 깎고 학습 데이터에 전혀 등장하지 않는 케이스들에는 작으나마 일부 확률을 부여하게 된다.

### 뉴럴 네트워크 기반 언어 모델

앞에서 설명한 **통계 기반 언어 모델**은 단어들의 빈도를 세어서 학습한다. 그런데 이를 뉴럴 네트워크로 학습할 수도 있다. 뉴럴 네트워크는 입력과 출력 사이의 관계를 유연하게 포착해낼 수 있고, 그 자체로 확률 모델로 가능할 수 있기 때문이다.

**뉴럴 네트워크 기반 언어 모델**은 주어진 단어 시퀸스를 가지고 다음 단어를 맞추는 과정에서 학습된다. 학습이 완료되면 이들 모델의 중간 혹은 말단 계산 결과물을 단어나 문장의 임베딩으로 활용한다. ELMo, GPT 등 모델이 여기에 해당한다.

**마스크 언어 모델**은 언어 모델 기반 기법과 큰 틀에서는 유사하지만 디테일에서 차이를 보이는 기법이다. 
- 발 없는 말이 [MASK] 간다. -> 언어 모델 -> [천리]

위와 같이 문장 중간에 마스크를 씌워 놓고, 해당 마스크 위치에 어떤 단어가 올지 예측하는 과정에서 학습한다.

**언어 모델 기반 기법**은 단어를 순차적으로 입력받아 다음 단어를 맞춰야 하기 때문에 태생적으로 **일방향**이다. 하지만 **마스크 언어 모델 기반 기법**은 문장 전체를 다 보고 중간에 있는 단어를 예측하기 때문에 **양방향**학습이 가능하다. 따라서 **마스크 언어 모델 기반 방법**이 기존 **언어 모델 기법들**대비 임베딩 품질이 좋다. **BERT**가 이 부류에 속한다.

## 지도 학습 기반 형태소 분석

품질 좋은 임베딩을 만들기 위해서는 문장이나 단어의 경계를 컴퓨터에게 알려줘야 한다. 그렇지 않으면 어휘 집합에 속한 단어 수가 기하급수적으로 늘어나서 연산의 비효율이 발생한다. 특히 한국어는 조사와 어미가 발달한 **교착어**이기 때문에 이러한 처리를 좀 더 섬세히 해줘야 한다.

예를 들어 한국어 동사 [가다]는 [가겠다, 가더라, 가겠더라, 가다] 등 여러 활용형 말뭉치도 존재한다면 어휘 집합이 무수히 많아질 수 있다. 이를 해결하기 위해 **형태소 분석 기법**을 사용한다.

[가겠다, 가더라]를 **온전한닢**이라는 오픈소스 형태소 분석기로 분석한 결과는 다음과 같다.

~~~
가겠다 -> 가, 겠, 다
가더라 -> 가, 더라
~~~

위와 같이 형태소를 분석한 뒤 각 형태소들을 단어로 취급한다면 어휘 집합은 총 4개가 된다. 여기에 [가겠더라]라는 활용형이 말충치에 추가됐다고 가정해보자. [가겠더라]는 [가, 겠, 더라]로 분석된다. 따라서 우리는 어휘 집합을 수정하지 않고도 활용형을 처리할 수 있게 된다.

따라서 교착어인 한국어는 한정된 종류의 조사와 어미를 자주 이용하기 때문에 각각에 대응하는 명사, 용언, 어간만 어휘 집합에 추가하면 취급 단어 개수가 많이 줄기 때문에 형태소 분석기만 잘 활용해도 자연어 처리의 효율성을 높일 수 있다.

이 절에서는 **지도 학습**기반의 형태소 분석 방법을 설명한다. 형태소 분석기들은 언어학 전문가들이 태깅한 형태소 분석 말뭉치로부터 학습된 지도 학습 기반 모델들이다.

<br/> - **태깅(tagging)** : 아래처럼 모델 입력과 출력 쌍을 만드는 작업을 가리킨다. 이 모델들은 문자열이 주어졌을 때 사람이 알려준 정답 패턴에 최대한 가깝게 토크나이즈한다.

~~~
입력 : 아버지가방에들어가신다
출력 : 아버지, 가, 방, 에, 들어가, 신다.
~~~
