# 2.4 어떤 단어가 같이 쓰였는가

## 2.4.1 분포 가정

* 자연어 처리에서 **분포**는 동시에 등장하는 이웃 단어 또는 문맥의 집합을 의미함.
* **분포가정**의 전제는 **어떠한 단어의 쌍이 비슷한 문맥에서 자주 등장한다면 그 의미 또한 유사할 것.** 

예문)
>저의 특기는 청소와 **빨래**입니다.  
뜨거운 물로 **빨래**를 하면 더 좋다.  
  
>찬 물로 옷을 **세탁**한다.  
**세탁**,청소,요리는 너무 힘들다

**타깃 단어**(빨래,세탁)는 **문맥 단어**(청소,물)가 비슷하기 때문에 타깃 단어들끼리 관계가 있을것이라고 생각할 수 있다.

## 2.4.2 분포와 의미 (1): 형태소

**형태소**: 의미를 가지는 최소 단위 (더 쪼개면 뜻을 잃어버린다)

예문)
> 철수가 밥을 먹었다.

철수와 밥은 더이상 쪼개면 뜻을 잃어버리게 된다.

* 언어학적으로 형태소를 분석하는 방법의 대표적인 예시로는 **계열 관계**가 있다.  

**계열 관계**:해당 형태소 자리를 다른 형태소로 대치할 수 있는가를 따지는 것 

>예를 들어 밥 대신 빵을 사용해도 되기 때문에 이때 **밥**에 형태소 자격을 부여하게 된다.

## 2.4.2 분포와 의미(2): 품사

**품사**: 단어를 문법적 성질의 공통성에 따라 묶어 놓은 것

품사 분류 기준은 **기능, 의미, 형식** 세가지로 나눌 수 있다.

예문)
> 이 샘의 **깊이**는 얼마야?  
저 산 의 **높이**가 얼마야?  
이 샘이 **깊다**.  
저 산이 **높다**.


 * **기능**: 한 단어가 문장 가운데서 다른 단어와 맺는 관계  
 
예문에서 깊이, 높이는 문장의 **주어** 깊다, 높다는 **서술어**로 사용이 되고 있다.

* **의미**: 단어의 형식적인 의미 
> (1) 깊이, 깊다  
(2) 높이 높다

위와 같이 묶는 방법을 어휘적 의미를 기준으로 묶었다고 하는데 품사 분류에서는 어휘적 의미보다는 형식적 의미가 중요하다.

* **형식**: 단어의 형태적 특징(사물의 이름을 나타내는지? 성질, 상태를 나타내는지?)  
> (1) 깊이, 높이  
(2) 깊다, 높다

위 예시처럼 깊이와 높이는 변화하지 않지만 깊다와 높다는 깊었다/높았다, 깊겠다/높겠다 등으로 변할 수 있다.

## 2.4.2 품사 분류의 어려움

>(a) 공부하다  
(b) 공부

(a)는 동사,(b)는 명사로 구분할 수 있다.여기서 공부라는 단어에 "하다"가 추가 되었다고 해서 그 의미 차이가 있다고는 결정 지어서 말할 수 없다. 이것이 의미가 품사 분류의 결정적인 기준이 될 수 없다는 이야기이다.

> (a) 영수가 학교에 간다.  
(b) 영수! 조용히 해.

여기서는 형태는 같지만 기능과 의미가 달라졌다.이처럼 형태도 품사 분류 시 결정적 기준이 될 수 없다.

한국어에서 품사를 분류할 때 가장 중요한 기준은 바로 **기능** 이라고 한다. 예시를 통해 설명을 하자면 (a), (b)에서는 (a)는 동작을 하는 기능이 있기 때문에 **동사** (b)는 사물의 이름을 나타내기 때문에 **명사**로 구분할 수 있다. 또한 (ㄱ)은 사물의 이름, (ㄴ)은 감탄사로 분류할 수 있다.

## 2.4.4 점별 상호 정보량

* **점별 상호 정보량**: 두 확률변수 사이의 상관성을 계량화하는 단위

확률수가 독립인 경우 값이 0이 되는데 이것은 단어A가 나타나는 것이 단어 B의 등장할 확률에 영향을 주지 않는다는 것이고 B도 그와 같다.

점별 상호 정보량(PMI)
\begin{align}
\log{\operatorname{P(A,B)}\over\operatorname{P(A)XP(B)}}
\end{align}

단어 A가 등장할때 단어 B와 같이 자주 나타나게되면?? PMI값은 커지게 된다.

* **단어 문맥 행렬**: PMI행렬의 행 벡터 자체를 단어의 임베딩으로 표현

위 예시처럼 window=4라고 한다면 타깃 단어를 기준으로 앞뒤로 4개의 문맥 단어의 빈도를 계산한다.

예를 들어서 전체 빈도 1000회 중에서 ***I***가 등장한 횟수가 20 ***to***가 등장한 횟수가 15, I와 to가 동시에 등장한 횟수를 10이라고 했을때 PMI를 계산해보면 다음과 같다

\begin{align}
 PMI(I,to) = \log{\operatorname{P(I,to)}\over\operatorname{P(I)XP(to)}}  
 =\log{\operatorname{\operatorname(10/1000)}\over\operatorname{(20/1000) x (15/1000)}}  
\end{align}

## 2.4.5 Word2Vec

분포가정의 대표적인 예시로 **같이 등장하는 횟수가 많을 수록 비슷한 의미를 가진다**는 것이 핵심 포인트다

**CBOW**


타깃 단어를 기준으로 앞뒤로 몇개의 단어를 Input으로 사용하여 타깃 단어를 유추하는 모델이다.

**Skip-gram**


**CBOW와 반대의 개념**으로 타깃 단어 하나를 가지고 주위에 등장하는 단어가 무엇일지 예측하는 것이다.
